```python
import matplotlib.pyplot as plt

# 기본스타일 적용용
plt.style.use("default")

# y값만 전달, x축은 자동으로 0, 1, 2, 3이 설정정
plt.plot([2, 4, 6, 8])
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [10, 30, 40, 76]
plt.plot(x, y)
plt.show()
```

```python
x = range(11)
y = range(15, 26)

plt.plot(x, y, "y.") # 색상옵션: b blue, g green, r red, c 청록색, m magenta, y yellow, k black, w white/ o 원, v 역삼각형, ^ 삼각형, s 사각형 + 플러스 . 점점
plt.show()
```

```python
fig = plt.figure()

a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(1, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, "y^")

x = range(0, 50)
y = range(5, 55)
a2.plot(x, y, "k+")
```

```python
import matplotlib 
import matplotlib.font_manager as fm

# 간격을 가능한 최적의 수치로 자동 조정정
#fig = plt.figure(constrained_layout=True)

# 원하는 퐅트 파일 경로 문자열로 입력
font_location = "C:\Windows\Fonts\PAPYRUS.TTF"

# 지정한 폰트 파일 경로에서 폰트 추출
fontname = fm.FontProperties(fname=font_location).get_name()

# matplotlib의 전역 폰트 설정 변경
matplotlib.rc('font', family = fontname)

fig = plt.figure()

a1 = fig.add_subplot(1, 2, 1)
a2 = fig.add_subplot(1, 2, 2)

x = range(0, 10)
y = range(5, 15)
a1.plot(x, y, "y^")

x = range(0, 50)
y = range(5, 55)
a2.plot(x, y, "k+")

# x축, y축 이름 지정
a1.set_xlabel("time")
a1.set_ylabel("US dollar")
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")


fig = plt.figure()

a = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]

# x축 라벨 이름(tick 이름)
labels = ['English', 'math']

# 각 막대에 적용할 컬러 설정
colors = ['red', 'blue']



a.bar(x, y, tick_label=labels, color = colors)

for i, v in enumerate(y) :
    a.text(v+2, i+1.05, str(v))

plt.show()
```
